# import dictionaries

```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sb
```

# read dataset

```python
data=pd.read_csv('google_review_ratings.csv')
```

```python
data
```

```python
data.isnull().sum()
```

# dropping unnecessary columns in the dataset

```python
my_data=data.drop(['Unnamed: 25','User'],axis=1)
```

```python
my_data
```

```python
my_data.isnull().sum()
```

```python
dt=my_data.dropna()
```

```python
dt
```

```python
dt.isnull().sum()
```

# importing Kmeans

```python
from sklearn.cluster import KMeans
```

```python
kmeans=KMeans(5)
kmeans.fit(dt)
```

```python
clusters=dt.copy()
clusters['cluster_pred']=kmeans.predict(dt)
clusters
```

```python
sb.scatterplot(x='Category 8',y='Category 10',data=clusters,hue='cluster_pred',palette='cool')
```

```python
from sklearn import preprocessing
x_scaled=preprocessing.scale(dt)
x_scaled
```

```python
a=[]
for i in range(1,30):
    kmeans=KMeans(i)
    kmeans.fit(x_scaled)
    a.append(kmeans.inertia_)
plt.plot(range(1,30),a,marker='+');
```

```python
km=KMeans(5)
km.fit(x_scaled)
cluster_new=dt.copy()
cluster_new['cluster_pred']=km.predict(x_scaled)
cluster_new
```

# HIERARICHAL_CLUSTERING

# importing normalize

```python
from sklearn.preprocessing import normalize
```

```python
scale=normalize(dt)
```

```python
scale
```

```python
scale=pd.DataFrame(scale,columns=dt.columns)
scale
```

```python
plt.figure(figsize=(10,10))
import scipy.cluster.hierarchy as cluster
dendogram=cluster.dendrogram(cluster.linkage(scale,method='ward'))
```

```python
plt.figure(figsize=(10,10))
import scipy.cluster.hierarchy as cluster
dendogram=cluster.dendrogram(cluster.linkage(scale,method='ward'))
plt.axhline(y=15.0,color='y',linestyle='-')
```

# No of clusters is found to be 4

```python
from sklearn.cluster import AgglomerativeClustering as ag
```

# importing Agglomerative

```python
cluste=ag(n_clusters=4)
```

```python
cluste.fit_predict(scale)
```
